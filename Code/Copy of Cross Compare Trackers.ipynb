### Code

```python
import cv2
import numpy as np
from ultralytics import YOLO
import os
import base64
from IPython.display import HTML, display
from moviepy.editor import VideoFileClip
import matplotlib
matplotlib.use('Agg')
from deep_sort_realtime.deepsort_tracker import DeepSort
# CORRECTED IMPORT: Use the installed yolox.tracker
from yolox.tracker.byte_tracker import BYTETracker
import gc
import pandas as pd

# Create a dummy args object that mimics argparse for BYTETracker
# This is a common pattern in the YOLOX/ByteTrack codebase
class ByteTrackArgs:
    def __init__(self,
                 track_thresh=0.6,
                 track_buffer=30,
                 match_thresh=0.8,
                 min_box_area=10,
                 mot20=False): # Add mot20 if needed, usually False
        self.track_thresh = track_thresh
        self.track_buffer = track_buffer
        self.match_thresh = match_thresh
        self.min_box_area = min_box_area
        self.mot20 = mot20

class VideoTracker:
    def __init__(self, model_path, pig_class_id=0, output_dir="tracked_videos_output", tracker_type="deepsort"):
        print("Initializing VideoTracker...")
        self.model = YOLO(model_path)
        self.pig_class_id = pig_class_id
        self.output_dir = output_dir
        os.makedirs(self.output_dir, exist_ok=True)
        print(f"Output directory set to: {self.output_dir}")

        self.tracker_type = tracker_type.lower()
        self.tracker = None

        if self.tracker_type == "deepsort":
            self.tracker = DeepSort(
                max_iou_distance=0.7,
                max_age=30,
                n_init=3,
                nn_budget=100
            )
            print("DeepSORT tracker loaded successfully!")
        elif self.tracker_type == "bytetrack":
            # CORRECTED BYTETracker INSTANTIATION:
            # Create an instance of the dummy args
            # You can customize these values
            byte_track_config = ByteTrackArgs(
                track_thresh=0.6,      # High confidence threshold
                track_buffer=30,       # Number of frames to keep lost tracks
                match_thresh=0.8,      # IoU matching threshold
                min_box_area=10        # Minimum box area to track
            )
            self.tracker = BYTETracker(byte_track_config)
            print("ByteTrack tracker loaded successfully!")
        else:
            raise ValueError(f"Unsupported tracker type: {tracker_type}. Choose 'deepsort' or 'bytetrack'.")

        print(f"YOLOv8 model and {self.tracker_type.upper()} tracker loaded successfully!")

    def process_video(self, video_path, display_gif=False):
        print(f"\n--- Processing video: {os.path.basename(video_path)} using {self.tracker_type.upper()} ---")
        cap = cv2.VideoCapture(video_path)

        if not cap.isOpened():
            print(f"Error: Could not open video file {video_path}")
            return

        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = int(cap.get(cv2.CAP_PROP_FPS))

        # BYTETracker from yolox.tracker.byte_tracker does not have a .frame_rate attribute
        # It's typically configured via the args object at initialization, or not needed directly
        # for update. We remove this line:
        # if self.tracker_type == "bytetrack":
        #    self.tracker.frame_rate = fps

        video_filename = os.path.basename(video_path)
        output_video_name = f"tracked_{self.tracker_type}_{os.path.splitext(video_filename)[0]}.mp4"
        output_video_path = os.path.join(self.output_dir, output_video_name)

        csv_filename = f"{os.path.splitext(video_filename)[0]}_{self.tracker_type}_tracking_data.csv"
        csv_path = os.path.join(self.output_dir, csv_filename)
        tracking_data = []

        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

        frame_count = 0
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            results = self.model.predict(source=frame, verbose=False)

            current_frame_detections_for_tracker = []

            # This `detection_confidences` dict was a workaround for DeepSORT's lack of track confidence
            # It's less directly used for ByteTrack's internal confidence reporting.
            # We'll keep it for consistency but the `conf_val` from track.score is preferred for ByteTrack.
            detection_confidences = {}

            for r in results:
                boxes = r.boxes
                for box_idx, box in enumerate(boxes):
                    x1, y1, x2, y2 = box.xyxy[0].tolist()
                    conf = box.conf[0].item()
                    cls = int(box.cls[0].item())

                    if cls == self.pig_class_id and conf > 0.5:
                        if self.tracker_type == "deepsort":
                            bbox_xywh = [x1, y1, x2 - x1, y2 - y1]
                            current_frame_detections_for_tracker.append((bbox_xywh, conf, cls))
                        elif self.tracker_type == "bytetrack":
                            # ByteTrack expects detections as a numpy array with (x1, y1, x2, y2, conf, cls)
                            current_frame_detections_for_tracker.append([x1, y1, x2, y2, conf, cls])

                        detection_confidences[(int(x1), int(y1), int(x2), int(y2))] = conf


            tracked_objects = []
            if self.tracker_type == "deepsort":
                tracked_objects = self.tracker.update_tracks(current_frame_detections_for_tracker, frame=frame)
            elif self.tracker_type == "bytetrack":
                if current_frame_detections_for_tracker:
                    dets = np.array(current_frame_detections_for_tracker, dtype=np.float32) # Ensure float32
                    # The update method typically takes detections, and optionally img_info, img_size
                    # For the BYTETracker from yolox, it often expects (dets, image_shape, ...)
                    # Let's check the signature from the original repo:
                    # update(self, output_results, img_info, img_size)
                    # output_results are detections in [x1,y1,x2,y2,score,cls] format
                    # img_info could be {'id': 0, 'height': H, 'width': W}
                    # img_size is (H, W)

                    # For now, let's pass the image shape as img_info and img_size if the method expects it
                    img_info = {'id': frame_count, 'height': frame_height, 'width': frame_width}
                    img_size = (frame_height, frame_width) # (height, width)

                    tracked_objects = self.tracker.update(dets, img_info, img_size)
                else:
                    # If no detections, pass an empty array, and relevant image info
                    img_info = {'id': frame_count, 'height': frame_height, 'width': frame_width}
                    img_size = (frame_height, frame_width)
                    tracked_objects = self.tracker.update(np.empty((0,6), dtype=np.float32), img_info, img_size)


            dark_blue_color = (255, 0, 0)
            bbox_thickness = 4
            text_thickness = 3

            for track in tracked_objects:
                track_id = -1
                x1, y1, x2, y2 = 0, 0, 0, 0
                conf_val = 0.0

                if self.tracker_type == "deepsort":
                    if not track.is_confirmed():
                        continue
                    track_id = track.track_id
                    ltrb = track.to_ltrb()
                    x1, y1, x2, y2 = int(ltrb[0]), int(ltrb[1]), int(ltrb[2]), int(ltrb[3])

                    conf_val = 1.0 # Placeholder for DeepSORT, as discussed

                elif self.tracker_type == "bytetrack":
                    # BYTETracker from yolox.tracker.byte_tracker returns `STrack` objects
                    # Check their state to ensure they are active tracks
                    # TrackState is an Enum, track.is_activated means it's confirmed
                    if not track.is_activated: # or track.state != STrack.TrackState.Tracked: (STrack needs to be imported)
                        continue

                    # Bounding box is tlwh (top-left, width, height)
                    x1, y1, w, h = track.tlwh
                    x2, y2 = x1 + w, y1 + h
                    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                    track_id = track.track_id
                    conf_val = track.score # This is the track's confidence (from the matched detection)

                if track_id != -1:
                    cv2.rectangle(frame, (x1, y1), (x2, y2), dark_blue_color, bbox_thickness)
                    text = f"ID: {track_id}"
                    if conf_val != 0.0: # Only display if it's not the initial 0.0 placeholder
                        text += f" C:{conf_val:.2f}"
                    cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, dark_blue_color, text_thickness)

                    tracking_data.append({
                        'frame_id': frame_count,
                        'object_id': int(track_id),
                        'bbox_x1': x1,
                        'bbox_y1': y1,
                        'bbox_x2': x2,
                        'bbox_y2': y2,
                        'confidence': conf_val
                    })

            out.write(frame)
            frame_count += 1

            if frame_count % 200 == 0:
                print(f"Processed {frame_count} frames for {os.path.basename(video_path)} using {self.tracker_type.upper()}")
                gc.collect()

        cap.release()
        out.release()

        print(f"Tracking complete for {os.path.basename(video_path)} using {self.tracker_type.upper()}! Output video saved to: {output_video_path}")

        if tracking_data:
            df = pd.DataFrame(tracking_data)
            df.to_csv(csv_path, index=False)
            print(f"Tracking data saved to: {csv_path}")
        else:
            print(f"No tracking data to save for {os.path.basename(video_path)} using {self.tracker_type.upper()}.")

        if display_gif:
            self._display_gif_from_video(output_video_path)

        del cap, out, frame, results, current_frame_detections_for_tracker, tracked_objects, tracking_data
        gc.collect()

    def _display_gif_from_video(self, video_path, gif_fps=10, resize_factor=0.5):
        gif_filename = f"{os.path.splitext(os.path.basename(video_path))[0]}.gif"
        gif_path = os.path.join(self.output_dir, gif_filename)

        if os.path.exists(video_path):
            try:
                with VideoFileClip(video_path) as clip:
                    clip = clip.set_fps(gif_fps)
                    clip = clip.resize(resize_factor)
                    clip.write_gif(gif_path, verbose=False)

                print(f"Video successfully converted to GIF: {gif_path}")

                with open(gif_path, "rb") as gif_file:
                    gif_data = gif_file.read()
                gif_b64 = base64.b64encode(gif_data).decode('utf-8')

                html_content = f"""
                <img src="data:image/gif;base64,{gif_b64}" width="100%" height="auto">
                """
                display(HTML(html_content))
            except Exception as e:
                print(f"Error converting video to GIF or displaying: {e}")
        else:
            print(f"Error: Video file not found at {video_path} for GIF conversion.")

# Example usage
video_files = [
    "/content/videos/Test Vid 1.mp4",
]

model_path = '/content/v9/weights/best.pt'
output_base_dir = "/content/tracked_videos"
pig_class_id = 0
```

```python
print("\n--- Running with DeepSORT ---")
tracker_deepsort = VideoTracker(
    model_path=model_path,
    pig_class_id=pig_class_id,
    output_dir=os.path.join(output_base_dir, "deepsort_results"),
    tracker_type="deepsort"
)
for video_file in video_files:
    tracker_deepsort.process_video(video_file, display_gif=False)
del tracker_deepsort
gc.collect()
```

```python
print("\n--- Running with ByteTrack ---")
tracker_bytetrack = VideoTracker(
    model_path=model_path,
    pig_class_id=pig_class_id,
    output_dir=os.path.join(output_base_dir, "bytetrack_results"),
    tracker_type="bytetrack"
)
for video_file in video_files:
    tracker_bytetrack.process_video(video_file, display_gif=False)
del tracker_bytetrack
gc.collect()
```
