```python
from ultralytics import YOLO

# Load a custom trained model
model = YOLO('/content/v9/weights/best.pt')

print("YOLOv8 model loaded successfully!")
```

# Task
Run YOLOv8 model on the video "/content/videos/Test Vid 1.mp4", save the output as a GIF, and display the GIF.

## Run inference

### Subtask:
Run the loaded YOLOv8 model on the specified video to detect pigs.


**Reasoning**:
Use the loaded YOLOv8 model to run inference on the specified video and save the output.



```python
results = model.predict(source="/content/videos/Test Vid 1.mp4", save=True)
```

```python
# Accessing bounding boxes for each object detected in each frame
for result in results:
    # Get the bounding boxes
    boxes = result.boxes

    # Iterate through each box
    for box in boxes:
        # Get the coordinates (x1, y1, x2, y2) of the bounding box
        x1, y1, x2, y2 = box.xyxy[0].tolist()

        # Get the confidence score
        conf = box.conf[0].item()

        # Get the class index
        cls = int(box.cls[0].item())

        # Print or use the bounding box information as needed
        print(f"Box coordinates: ({x1}, {y1}) ({x2}, {y2}), Confidence: {conf}, Class: {cls}")
```

## Convert video to gif

### Subtask:
Convert the saved video with detections into a GIF format.


**Reasoning**:
Convert the saved video with detections into a GIF format using moviepy.



```python
from moviepy.editor import VideoFileClip
import os

video_path = "runs/detect/predict/Test Vid 1.avi"
video_path = "/content/tracked_output.mp4"
gif_path = "Test_Vid_1_output.gif"

if os.path.exists(video_path):
    with VideoFileClip(video_path) as clip:
        clip = clip.set_fps(10)
        clip = clip.resize((clip.w // 2, clip.h // 2))
        clip.write_gif(gif_path)

    print(f"Video successfully converted to GIF: {gif_path}")
else:
    print(f"Error: Video file not found at {video_path}")
```

## Display gif

### Subtask:
Display the generated GIF in the notebook using HTML.


**Reasoning**:
Display the generated GIF in the notebook using HTML.



```python
import base64
from IPython.display import HTML, display

gif_path = "Test_Vid_1_output.gif"

with open(gif_path, "rb") as gif_file:
    gif_data = gif_file.read()
gif_b64 = base64.b64encode(gif_data).decode('utf-8')

html_content = f"""
<img src="data:image/gif;base64,{gif_b64}" width="100%" height="auto">
"""

display(HTML(html_content))
```
